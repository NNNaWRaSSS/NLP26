```python
from scipy import spatial
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
cosine_similarity= lambda vec1,vec2: 1-spatial.distance.cosine(vec1,vec2)
```

```python
import spacy
import spacy.cli

# The previously loaded nlp model (en_core_web_sm) does not have word vectors.
# To use word vectors, a larger model like 'en_core_web_lg' is required.
try:
    nlp_with_vectors = spacy.load("en_core_web_lg")
except OSError:
    print("Downloading 'en_core_web_lg' model...")
    spacy.cli.download("en_core_web_lg")
    nlp_with_vectors = spacy.load("en_core_web_lg")

boy=nlp_with_vectors.vocab['boy'].vector
brother=nlp_with_vectors.vocab['brother'].vector
sister=nlp_with_vectors.vocab['sister'].vector
```

```python
new_vector=(brother-boy)+sister
```

```python
computed_similarities=[]

for word in nlp.vocab:
    if word.has_vector and word.is_lower and word.is_alpha and not word.is_stop:
       # if word not in stop_words:
        similarity=cosine_similarity(new_vector,word.vector)
        if similarity>=0.1:
            computed_similarities.append((similarity,word.text))
```

```python
computed_similarities=sorted(computed_similarities,reverse=True)
```

```python
print(computed_similarities)
```
